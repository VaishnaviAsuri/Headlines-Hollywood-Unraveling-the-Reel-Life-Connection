```python
import pandas as pd
from omdbapi.movie_search import GetMovie
from fuzzywuzzy import fuzz, process  # Install fuzzywuzzy using: pip install fuzzywuzzy

# Initialize an empty list to store release dates
release_dates = []

# Replace 'your_api_key' with your actual OMDB API key
api_key = '172e4e94'

# Create an instance of GetMovie with your API key
movie_api = GetMovie(api_key=api_key)

# Iterate through 'Movie_Title' column and query the API
for movie_title in df1['Movie_Title']:
    # Use try-except to handle potential API request errors
    try:
        # Use fuzzy string matching to find the closest matching movie title
        search_results = process.extractOne(movie_title, df1['Movie_Title'], scorer=fuzz.token_sort_ratio)
        
        if search_results[1] > 80:  # Adjust the threshold as needed
            closest_match = search_results[0]
            movie = movie_api.get_movie(title=closest_match)
            released_date = movie.get('released', 'N/A')
            release_dates.append(released_date)
        else:
            release_dates.append('N/A')
    except Exception as e:
        print(f"Error fetching data for '{movie_title}': {e}")
        release_dates.append('N/A')

# Add the release dates as a new column in the DataFrame
df1.loc[:, 'Release_Date'] = release_dates

# Save the updated DataFrame to a new CSV file
output_csv_file = 'movies_with_release_dates1.csv'
df1.to_csv(output_csv_file, index=False)

print(f"Release dates retrieved and saved to '{output_csv_file}'.")
```

```python
import pandas as pd
from omdbapi.movie_search import GetMovie
from fuzzywuzzy import fuzz, process  # Install fuzzywuzzy using: pip install fuzzywuzzy

# Initialize empty lists to store Director, Actors, Year, and Runtime
directors = []
actors = []
years = []
runtimes = []

# Replace 'your_api_key' with your actual OMDB API key
api_key = '1a6b3e56'

# Create an instance of GetMovie with your API key
movie_api = GetMovie(api_key=api_key)

# Iterate through 'Title' column in the dff1 DataFrame and query the API
for movie_title in ddf1['Title']:
    # Use try-except to handle potential API request errors
    try:
        # Use fuzzy string matching to find the closest matching movie title
        search_results = process.extractOne(movie_title, ddf1['Title'], scorer=fuzz.token_sort_ratio)
        
        if search_results[1] > 80:  # Adjust the threshold as needed
            closest_match = search_results[0]
            movie = movie_api.get_movie(title=closest_match)
            
            # Extract Director, Actors, Year, and Runtime
            director = movie.get('director', 'N/A')
            actor = movie.get('actors', 'N/A')
            year = movie.get('year', 'N/A')
            runtime = movie.get('runtime', 'N/A')
            
            directors.append(director)
            actors.append(actor)
            years.append(year)
            runtimes.append(runtime)
        else:
            directors.append('N/A')
            actors.append('N/A')
            years.append('N/A')
            runtimes.append('N/A')
    except Exception as e:
        print(f"Error fetching data for '{movie_title}': {e}")
        directors.append('N/A')
        actors.append('N/A')
        years.append('N/A')
        runtimes.append('N/A')

# Add the Director, Actors, Year, and Runtime as new columns in the dff1 DataFrame
ddf1['Director'] = directors
ddf1['Actors'] = actors
ddf1['Year'] = years
ddf1['Runtime'] = runtimes

# Save the updated DataFrame to a new CSV file
output_csv_file = 'ddf1_with_movie_details.csv'
ddf1.to_csv(output_csv_file, index=False)

print(f"Movie details retrieved and saved to '{output_csv_file}'.")
```
